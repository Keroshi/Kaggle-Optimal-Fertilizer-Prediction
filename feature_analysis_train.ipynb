```python
import os
import pandas as pd

file_path = "./inputs/"
files = os.listdir(file_path)
files
```

```python
# Load the data
df = pd.read_csv(file_path + files[2]).set_index("id")
df
```

```python
df.describe()
```

```python
# Missing Values?
df.isnull().sum()
```

```python
# Determine the categorical and non-categorical columns
categories = df.select_dtypes(include=object).columns[:-1]
non_categories = df.select_dtypes(include=["number"]).columns

display(categories)
display(non_categories)
```

```python
# Unique in the target column
df["Fertilizer Name"].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in non_categories:
    plt.figure(figsize=(15, 8))
    plt.subplot(1, 2, 1)
    sns.boxplot(data=df, y=column, )
    plt.subplot(1, 2, 2)
    sns.histplot(data=df, x=column, kde=False)
    plt.show()
```

```python
for column in categories:
    plt.figure(figsize=(15, 8))
    sns.countplot(data=df, x=column, order=df[column].value_counts().index)
    plt.title(f'Countplot of {column}')
    plt.xticks(rotation=45)
    plt.show()
```
