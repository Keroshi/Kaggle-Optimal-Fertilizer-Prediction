```python
import os
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.metrics import average_precision_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

file_path = "./inputs/"
files = os.listdir(file_path)
files
```

```python
train_df = pd.read_csv(file_path + files[2]).set_index("id")
target = train_df["Fertilizer Name"]
train_df.drop("Fertilizer Name", axis=1, inplace=True)
train_df
```

```python
# Filter the categorical and non-categorical columns
categories = train_df.select_dtypes(include=object).columns
non_categories = train_df.select_dtypes(include=["number"]).columns

display(categories)
display(non_categories)
```

```python
# Scaling the non_categorical features
scaler = StandardScaler()
train_df[non_categories] = scaler.fit_transform(train_df[non_categories])
train_df[non_categories]
```

```python
# Encode the categorical features
encoder = LabelEncoder()
for column in categories:
    train_df[column] = encoder.fit_transform(train_df[column])

train_df[categories]
```

```python
train_df
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(train_df, target, test_size=0.2, random_state=42)
```

```python
# First training
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
score = average_precision_score(y_test, predictions)
display(score)
display(predictions)
```

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
score = average_precision_score(y_test, predictions)
display(score)
display(predictions)
```
